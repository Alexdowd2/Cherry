```python
import pandas as pd
import matplotlib
import matplotlib.style as style
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"O:\Procurement Planning\Tableau\Alex Tableau\model_output.csv")
```

```python
df = df[['candidate_inc','candidate_chal','ecwin_inc','ecwin_chal','modeldate']]
```

```python
df['ecwin_chal pct'] = df['ecwin_chal']*100
df['ecwin_inc pct'] = df['ecwin_inc']*100
```

```python
style.use('fivethirtyeight')

fig,ax = plt.subplots()
ax.plot('modeldate','ecwin_inc pct',color='red',data=df,label='Trump',linewidth=10)
ax.plot('modeldate','ecwin_chal pct',color='blue',data=df,label='Biden',linewidth=10)
ax.axes.get_xaxis().set_visible(False)

fig.set_figheight(10)
fig.set_figwidth(20)

plt.legend(loc=5,prop={'size':25})
plt.title('Chance of Winning Electoral College',fontweight='bold',fontsize='25')
plt.show()
```
